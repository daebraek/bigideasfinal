For my submetric, I looked at the avarage daily car traffic by neighborhood using Pittsburghs traffic counter dataset.

For my dataset the less traffic a city has the better, this is because you can get around more easily and there is less noise compared to somewhere with a lot of traffic.

```python
import pandas as traffic

traffic = traffic.read_csv("traffic_data.csv")
traffic.head()
```

To measure this I grouped the data by neighborhood and calculated the mean avarage daily traffic of each one. Then I printed out the ten neighborhoods with the least amount of traffic.

```python
#group the avarage daily traffic by neighborhood avg
traffic_mean = traffic.groupby("neighborhood")["average_daily_car_traffic"].mean()
#sort by the 10 lowest traffic neighborhoods
lowest_traffic_mean = traffic_mean.sort_values().head(10)
#print them
print(lowest_traffic_mean)
```

The I created a chart for visibility

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(lowest_traffic_mean.index, lowest_traffic_mean.values, color='steelblue')
plt.xlabel('Average Daily Car Traffic')
plt.title('Top 10 Lowest Traffic Neighborhoods in Pittsburgh')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

According to this data, South Side Flats have the least amount of traffic
